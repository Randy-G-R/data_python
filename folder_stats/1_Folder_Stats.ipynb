```python
'''Analyzes a folder in your computer and shows how many 
programming scripts of different kind it has'''

import os
import matplotlib.pyplot as plt
```

```python
# Write the path of the folder to analize

base_folder = r'C:\'
```

```python
'''In this example I will analyze how many Python, Cpp
HTML, CSS, JS and JSON files there are in a folder where I
save my programming projects'''

# Count each type of file (Python, Cpp, HTML, CSS, JS)

python_files = 0
cpp_files = 0
html_files = 0
css_files = 0
js_files = 0
json_files = 0


for root, dirs, files in os.walk(base_folder):
    for file in files:
        if file.endswith('py'): python_files += 1
        if file.endswith('cpp'): cpp_files += 1
        if file.endswith('html'): html_files += 1
        if file.endswith('css'): css_files += 1
        if file.endswith('js'): js_files += 1
        if file.endswith('json'): json_files += 1

print(f'Number of python files: {python_files}')
print(f'Number of cpp files: {cpp_files}') 
print(f'Number of html files: {html_files}')
print(f'Number of css files: {css_files}')
print(f'Number of js files: {js_files}')
print(f'Number of json files: {json_files}')

total_files = python_files + cpp_files + html_files + css_files + js_files

# Calculate the percentaje of each type of file in the total

perc_python  =  python_files/total_files * 100
perc_cpp =  cpp_files/total_files * 100
perc_html =  html_files/total_files * 100
perc_css =  css_files/total_files * 100
perc_js =  js_files/total_files * 100
perc_json =  json_files/total_files * 100

print('---------------------------------------------')
print('PERCENTAGES:')
print()
print(f'Python: {perc_python}')
print(f'C++: {perc_cpp}')
print(f'HTML: {perc_html}')
print(f'CSS: {perc_css}')
print(f'JavaScript: {perc_js}')
print(f'JSON: {perc_json}')
```

```python
# MATPLOTLIB GRAPHS

labels = ['html_files', 'python_files', 'css_files', 'js_files', 'cpp_files', 'json_files']

data = [html_files, python_files, css_files, js_files, cpp_files, json_files]

# BAR GRAPH

plt.bar(labels, data)
plt.show
```

```python
# PIE CHART 

plt.pie(data, labels=labels)
plt.show()
```
